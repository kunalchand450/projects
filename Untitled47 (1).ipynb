```python
wait = WebDriverWait(driver,10)
Urls=[]
for i in range(0,2): #increase the range will increase the range quatity
    Url_links=driver.find_elements_by_xpath("//li[@class='detail_title']/a")


    for i in Url_links:
        
        Urls.append(i.get_attribute("href"))
   
    nxt_button=driver.find_element_by_xpath('//*[@id="search-result-container"]/div[4]/div/div/div[2]/nav/div[2]/ul/li[4]/a')
    nxt_button.click()
    time.sleep(5)
    
```

```python
len(Urls)

```

```python
Person_Names=[]
Bar_No=[]
Profile_link=[]
Title=[]
Company_name=[]
Address=[]
Phone_No=[]
Cell=[]
Fax=[]
#Email=[]
Website=[]
Board_Certification=[]
Bar_No=[]
#Email_Validation=[]


for i in Urls:
    driver.get(i)
    time.sleep(3)
    try:
        name=driver.find_element_by_xpath('//h1[@class="profile-title--bold"]')
        Person_Names.append(name.text.replace("\n"," "))
    except:
        Person_Names.append("-")
        
    try:    
        title = driver.find_element_by_xpath('//li[@class="masthead-list__item masthead-list__item--bold"]') 
        Title.append(title.text.replace("\n"," "))
    except:
        Title.append("-")
        
    try:   
        address=driver.find_element_by_xpath('/html/body/div[2]/div/div[3]/section[1]/div[2]/div[2]/ul/li[2]/address') 
    
        Address.append(address.text.replace("\n"," "))
    except:
        Address.append("-")
        
    try:    
        website = driver.find_element_by_xpath('//a[@class="view-website webstats-website-click"]') 
        Website.append(website.get_attribute("href"))
    except:
        Website.append("-")
        
    company = driver.find_elements_by_xpath('//a[@class="link link--gray"]')
    for i in company:
        try:
            Company_name.append(i.text.replace("\n"," "))
        except:
            Company_name.append("-")
    try:    
        phone=driver.find_element_by_xpath('//*[@id="education-section"]/div/div/div[1]/div[2]/span[1]') 
        Phone_No.append(phone.text.replace("\n"," ")) 
        
             
    except:
            Phone_No.append("-")
        
    try:
        board_certification = driver.find_element_by_xpath('//div[@class="small-12 medium-9 columns experience-value more-value"]')
        Board_Certification.append(board_certification.text.replace("\n"," "))
    except:
        Board_Certification.append("No Certification")
            
    cell = driver.find_elements_by_xpath('//*[@id="education-section"]/div/div/div[1]/div[2]/span[1]')
    for i in cell:
        try:
            Cell.append(i.text.replace("\n"," "))
        except:
            Cell.append("-")
    try:
            fax=driver.find_element_by_xpath('//*[@id="education-section"]/div/div/div[1]/div[2]/span[3]')
    
            Fax.append(fax.text.replace("\n"," "))   
    except:
            Fax.append("-")
    
    try:
            bar_no=driver.find_element_by_xpath('//*[@id="education-section"]/div/div/div[8]')
    
            Bar_No.append(bar_no.text.replace("ISLN:\n",""))   
    except:
            Bar_No.append("-")
```

```python
Company_Name=[] #will eleminate the extra items and produse a more refine list
for i in range(0,len(Company_name),2):
    Company_Name.append(Company_name[i].strip())
```

```python
import pandas as pd
df1=pd.DataFrame({})
df1['Person Name']=Person_Names
df1['Profile link']=Urls
df1['Title']=Title
df1['Address']=Address
df1['Website']=Website
df1['Company Name']=Company_Name
df1['Phone No']=Phone_No
df1['Board Certification']=Board_Certification
df1['Cell']=Cell
df1['Fax']=Fax
df1["Bar No"]=Bar_No
ATTORNEY_PROFILE=df1
```

```python
# writing to Excel
datatoexcel2 = pd.ExcelWriter('ATTORNEY_PROFILE.xlsx')
# write DataFrame to excel
ATTORNEY_PROFILE.to_excel(datatoexcel2)
# save the excel
datatoexcel2.save()
print('DataFrame2 is written to Excel File successfully.')
```
